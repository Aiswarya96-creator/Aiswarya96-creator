data transformation

```python
gdp.rename(columns={'Country':'country'},inplace=True)
gdp
```

```python
litracy.rename(columns={'name':'country'},inplace=True)
litracy
```

```python
litracy['literacy_rate']=pd.to_numeric(litracy['literacy_rate'])
litracy
```

```python
population['population']=pd.to_numeric(population['population'])
population
```

```python
users['internet_users']=pd.to_numeric(users['internet_users'])
users
```

```python
gdp['GDP']=pd.to_numeric(gdp['GDP'])
gdp
```

```python
merged=population.merge(gdp,on="country")
merged=merged.merge(users,on="country")
merged=merged.merge(litracy,on="country")
merged
```

```python
merged["Internet Penetration (%)"] = (
    merged["internet_users"] /
    merged["population"] *100
).round(2)

merged["GDP Per Capita"] = (
    merged["GDP"] /
    merged["population"]
).round(2)

merged.head()
```

load data

```python
conn = sqlite3.connect("country_data.db")

merged.to_sql(
    "country_statistics",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT *
FROM country_statistics
LIMIT 5;
"""

pd.read_sql(query,conn)
```

ETL process explaination
data pipeline that ingested information from CSV, Excel, JSON, and XML files, cleaned and standardized the data using Python's Pandas library, integrated the datasets by country, and calculated new metrics for Internet Penetration and GDP Per Capita before storing the final result in a SQLite database. Subsequent analysis revealed that the United States holds the highest GDP, China leads in both population and internet user count, the average literacy rate across the studied countries is 97.248%, and there is a strong positive correlation of approximately 0.773 between literacy rates and internet penetration.
